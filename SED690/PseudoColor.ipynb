Pseudo colors

Cloud_8Feb2025.jpg

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from PIL import Image
import matplotlib.pylab as plt
```

```python


image_bgr = io.imread("/content/Cloud_8Feb2025.jpg")   # use the above image or any of your preferred image.
img = cv.resize(image_bgr, (0,0), fx=0.5, fy=0.5)
# Check width, height, and dimension
cv2_imshow(img)
print(img.shape)
print("Grayscale")

```

```python
# Prepare for 3 channels
threeChannel = np.ones((img.shape[0],img.shape[1],3), dtype=np.uint8) # prepare the storage with the same size as the original image

threeChannel[:,:,0] *= img[:,:]  # B
threeChannel[:,:,1] *= img[:,:]  # G
threeChannel[:,:,2] *= 255  # R

# image_oneChannel = np.array(oneChannel, np.uint8)
cv2_imshow(threeChannel)
print("Pseudo color", threeChannel.shape)
```
